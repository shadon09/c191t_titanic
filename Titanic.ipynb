## Load in train and test sets
## Make sure files are loading by printing out entries

```python
# Import test and train datasets
df_train = pd.read_csv('train.csv')
df_test = pd.read_csv('test.csv')

# View first lines of training data
print (df_train.info())
df_train.head(n=418)
```

## Set all the people that survived to ones that do not have any siblings or parents
## They don't have anyone to care about on board, and take up less space on emergency rafts, so they have a better chance of survival
## Print out first four entries as text

```python
df_test['Survived'] = (df_test.SibSp == 0) & (df_test.Parch == 0)
print (df_test.head())
```

## Print out data

```python
df_test['Survived'] = df_test.Survived.apply(lambda x: int(x))
df_test.head()
```

## Put data in csv file

```python
df_test[['PassengerId', 'Survived']].to_csv('loners_survive.csv', index=False)
print (df_test[['PassengerId', 'Survived']].head())
```
